```python
import pandas as  pd
import openpyxl
```

```python
df = pd.read_excel('2023-2UNLOCODECodeList.xlsx')
```

```python
df1 = df.copy()
df1.dropna(subset=['Coordinates'], inplace=True)
df1['slat'] = df['Coordinates'].str.split().str[0]
df1['slong'] = df['Coordinates'].str.split().str[1]
df1['lat'] = df1['slat'].str.extract('(\d+)').astype(int)/100
df1.loc[df1['slat'].str.endswith('S'), 'lat'] *= -1
df1['long'] = df1['slong'].str.extract('(\d+)').astype(int)/100
df1.loc[df1['slong'].str.endswith('W'), 'long'] *= -1

df1
```

```python
vLat = -14.0
vLong = 12.0
diff = 2.0
hiLat = vLat + diff
loLat = vLat - diff
hiLong = vLong + diff
loLong = vLong - diff
hiLat = 90 if hiLat > 90 else vLat + diff
lolat = -90 if loLat < -90 else vLat - diff
hiLong = 360-hiLong if hiLong > 180 else vLong + diff
loLong = 360-loLong if loLong < -180 else vLong - diff
print(vLat, hiLat, lolat, vLong, hiLong, loLong)
```
